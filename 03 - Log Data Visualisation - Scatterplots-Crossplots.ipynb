```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio as ls
```

```python
las = ls.read('data/15_9-19_SR_COMP_.las')
```

Within petrophysics scatterplots, or crossplots, are routinely used as part of the interpretation workflow. They allow us to determine key interpretation parameters such as </br>

- clay and shale end points for our clay or shale volume calculations</br>
- outlier detection</br>
- lithology identification</br>
- hydrocarbon identification</br>
- rock typing</br>
- regression analysis</br>
- and more</br>

```python
well = las.df()
```

```python
well.describe()
```

```python
well.head(10)
```

```python
well.plot(y='DEN')
well.plot(y='GR')
well.plot(y='NEU')
```

# Creating a Crossplot / Scatterplot

```python
# set up the scatter plot
plt.scatter(x='NEU', y='DEN', data=well)
plt.show()
```

```python
# set the default plot size for our scatterplots
plt.rcParams['figure.figsize'] = (8,8)
```

```python
plt.scatter(x='NEU', y='DEN', data=well)
plt.xlim(-5,60)     # range of values for 'NEU'
plt.ylim(3.0,1.5)   # range of values for 'DEN', values in reverse order
plt.show()
```

Adding Labels to the Axes

```python
plt.scatter(x='NEU', y='DEN', data=well)
plt.xlim(-5,60)
plt.ylim(3.0,1.5)
plt.xlabel('Neutron Porosity (NEU) - %', fontsize=14)
plt.ylabel('Bulk Density (DEN) - g/cc', fontsize=14)
plt.show()
```

We can add a third variable onto our scatterplot through the use of colour. For this plot, we will add in the c argument and pass it the Gamma Ray (GR) column from the dataframe. </br>

To control the range of colours shown we need to pass in values to vmin and vmax. In this example, we will set these to 0 and 100.

```python
plt.scatter(x='NEU', y='DEN', data=well, c='GR', vmin=0, vmax=100)
plt.xlim(-5,60)
plt.ylim(3.0,1.5)

plt.xlabel('Neutron Porosity (NEU) - %', fontsize=14)
plt.ylabel('Bulk Density (DEN) - g/cc', fontsize=14)

plt.show()
```

# Changing Colormap and Adding Colorbar </br>
https://matplotlib.org/stable/tutorials/colors/colormaps.html </br>

Style sheets allow us to control the look and feel of the plots.</br>
https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html 

```python
# Set up the scatter plot
plt.scatter(x='NEU', y='DEN', data=well, c='GR', vmin=0, vmax=100, cmap='rainbow')

# Change the X and Y ranges
plt.xlim(-5,60)     # Range of values for 'NEU'
plt.ylim(3.0,1.5)   # range of values for 'DEN', values in reverse order

plt.xlabel('Neutron Porosity (NEU) - %', fontsize=14)
plt.ylabel('Bulk Density (DEN) - g/cc', fontsize=14)

# Make the colorbar show
plt.colorbar(label='Gamma Ray - API')

#Set the style sheet to bmh
plt.style.use('bmh') # Adding Gridlines & Plot Styling

plt.savefig('images/scatterPlot.png')
plt.show()
```

# Changing the Data
