```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio as ls
```

```python
las = ls.read('data/15_9-19_SR_COMP_.las')
```

```python
well = las.df()
```

```python
well.describe()
```

# Creating Histograms

```python
plt.hist(well['GR'])
plt.show()
```

```python
plt.hist(well['GR'], bins=30)
plt.show()
```

```python
plt.hist(well['GR'], bins=30, edgecolor='black')
plt.show()
```

```python
plt.hist(well['GR'], bins=30, edgecolor='black', color='red', alpha=0.5)
plt.xlabel('Gamma Ray - API', fontsize=14)
plt.ylabel('Frecuency', fontsize=14)
plt.xlim(0,200)

plt.savefig('images/histogram.png', dpi=300)

plt.show()
```

```python
well['GR'].plot(kind='hist', bins=30, edgecolor='black', color='red', alpha=0.5, density=True)
well['GR'].plot(kind='kde', color='black')
plt.xlabel('Gamma Ray - API', fontsize=14)
plt.ylabel('Frecuency', fontsize=14)
plt.xlim(0,200)
plt.show()
```

When calculating clay and shale volumes we often use the percentiles as our interpretation parameters.

```python
mean = well['GR'].mean()
p5 = well['GR'].quantile(0.05)
p95 = well['GR'].quantile(0.95)

print(f'Mean:\t {mean}')
print(f'P05:\t {p5}')
print(f'P95:\t {p95}')
```

```python
plt.hist(well['GR'], bins=30, edgecolor='black', color='red', alpha=0.5)
plt.xlabel('Gamma Ray - API', fontsize=14)
plt.ylabel('Frecuency', fontsize=14)
plt.xlim(0,200)

plt.axvline(mean, color='blue', label='mean')
plt.axvline(p5, color='yellow', label='5th Percentile')
plt.axvline(p95, color='green', label='95th Percentile')

plt.legend()
plt.savefig('images/histogram.png')
plt.show()
```
