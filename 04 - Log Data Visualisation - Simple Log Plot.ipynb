```python
import pandas as pd
import lasio as ls
import matplotlib.pyplot as plt
```

```python
lasio = ls.read('data/15_9-19_SR_COMP_.las')
```

```python
df = lasio.df()
```

```python
df.head()
```

To make it easier to work with our dataframe, we can convert the dataframe index, which is set to depth, to a column within the dataframe. We can achieve this by reseting the index like so.

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df. rename(columns={'DEPT':'DEPTH'}, inplace=True)
```

# Creating a Simple Line Plot

```python
df.plot('GR','DEPTH')
```

# Quick Subplot
If we want to view all of the columns within the dataframe, we can generate a subplot grid. </br>

This is done by taking the same line as before (df.plot()), and instead of passing in curve names, we pass in subplots=True.</br> We can also specify a figure size (figsize(), which controls how large the plot will appear.

```python
df.plot(subplots=True, figsize=(15,15))
```

# Working with Subplots in Matplotlib

```python
# Crear una figura (el lienzo para los gráficos) y un conjunto de ejes (región para trazar)
# figsize=(10, 10) establece el tamaño de la figura en 10x10 pulgadas
fig = plt.subplots(figsize=(10,10))

# Crear un conjunto de ejes dentro de la figura utilizando la función subplot2grid
# (1, 1) especifica que habrá 1 fila y 1 columna en la cuadrícula de ejes
# (0, 0) establece la posición inicial del eje en la esquina superior izquierda
# rowspan=1 y colspan=1 indican que el eje ocupará 1 fila y 1 columna
ax1 = plt.subplot2grid((1,1), (0,0), rowspan=1, colspan=1)

ax1.plot('GR', 'DEPTH', data=df, color='green') # Call the data from the df dataframe
ax1.set_xlabel('Gamma')
ax1.set_xlim(0,200)
ax1.set_ylim(4700,3500)
ax1.grid() # Display the grid
```

# Adding the Resistivity Log

```python
fig = plt.subplots(figsize = (10,10))

# (1, 2) especifica que habrá 1 fila y 2 columnas en la cuadrícula de ejes :    ax1 y ax2
# (0, 0) establece la posición inicial del primer eje en la primera columna :   ax1
# (0, 1) establece la posición inicial del segundo eje en la segunda columna:   ax2
ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,2), (0,1), rowspan=1, colspan=1)

ax1.plot('GR', 'DEPTH', data = df, color = 'green')
ax1.set_xlabel('Gamma')
ax1.set_xlim(0,200)
ax1.set_ylim(4700,3500)
ax1.grid()

ax2.plot('RDEP', 'DEPTH', data =df, color = 'red')
ax2.set_xlabel('Deep Res.')
ax2.set_xlim(0.2,2000)
ax2.semilogx()
ax2.set_ylim(4700,3500)
ax2.grid()
```

# Adding the Density Log

```python
fig = plt.subplots(figsize = (10,10))

# (1, 3) especifica que habrá 1 fila y 3 columnas en la cuadrícula de ejes :    ax1, ax2 y x3
# (0, 0) establece la posición inicial del primer eje en la primera columna :   ax1
# (0, 1) establece la posición inicial del segundo eje en la segunda columna:   ax2
# (0, 2) establece la posición inicial del tercer eje en la tercera columna:   ax3
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)

ax1.plot('GR', 'DEPTH', data = df, color = 'green')
ax1.set_xlabel('Gamma')
ax1.set_xlim(0,200)
ax1.set_ylim(4700,3500)
ax1.grid()

ax2.plot('RDEP', 'DEPTH', data = df, color = 'red')
ax2.set_xlabel('Deep Res.')
ax2.set_xlim(0.2,2000)
ax2.semilogx()
ax2.set_ylim(4700,3500)
ax2.grid()

ax3.plot('DEN', 'DEPTH', data = df, color = 'red' )
ax3.set_xlabel('Density')
ax3.set_xlim(1.95,2.95)
ax3.set_ylim(4700,3500)
ax3.grid()
```

# Ordenando elementos comunes

```python
fig, axes = plt.subplots(figsize = (10,10))

curve_names = ['Gamma', 'Deep Res.', 'Density']

ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)

ax1.plot('GR', 'DEPTH', data = df, color = 'green')
ax1.set_xlim(0,200)

ax2.plot('RDEP', 'DEPTH', data = df, color = 'red')
ax2.set_xlim(0.2,2000)
ax2.semilogx()

ax3.plot('DEN', 'DEPTH', data = df, color = 'red' )
ax3.set_xlim(1.95,2.95)

for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)
    ax.grid()
    ax.set_xlabel(curve_names[i])
```

```python
fig, axes = plt.subplots(figsize = (10,10))

curve_names = ['Gamma', 'Deep Res.', 'Density']

ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)

ax1.plot('GR', 'DEPTH', data = df, color = 'green')
ax1.set_xlim(0,200)

ax2.plot('RDEP', 'DEPTH', data = df, color = 'red')
ax2.set_xlim(0.2,2000)
ax2.semilogx()

ax3.plot('DEN', 'DEPTH', data = df, color = 'red' )
ax3.set_xlim(1.95,2.95)

for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)
    ax.grid()
    ax.set_xlabel(curve_names[i])

# To tidy the plot up a little more, we can remove the depth labels in between each subplot / track and reduce the space between them.
#Hide tick labels on the y-axis 
for ax in [ax2,ax3]:
    plt.setp(ax.get_yticklabels(), visible=False)

#Reduce the space between each subplot
fig.subplots_adjust(wspace = 0.05)
```

```python
fig, axes = plt.subplots(figsize = (10,10))

curve_names = ['Gamma', 'Deep Res.', 'Density', 'Neutron']

ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)
ax4 = ax3.twiny()

ax1.plot('GR', 'DEPTH', data = df, color = 'green', lw=0.5)
ax1.set_xlim(0,200)
ax1.spines['top'].set_edgecolor('green')

ax2.plot('RDEP', 'DEPTH', data = df, color = 'red', lw=0.5)
ax2.set_xlim(0.2,2000)
ax2.semilogx()
ax2.spines['top'].set_edgecolor('red')

ax3.plot('DEN', 'DEPTH', data = df, color = 'red', lw=0.5 )
ax3.set_xlim(1.95,2.95)
ax3.spines['top'].set_edgecolor('red')

ax4.plot('NEU', 'DEPTH', data=df, color='blue', lw=0.5)
ax4.set_xlim(45,-15)
ax4.spines['top'].set_edgecolor('blue')

for i, ax in enumerate(fig.axes):
    ax.set_ylim(4700,3500)

    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    ax.set_xlabel(curve_names[i])

    if i == 3:
        ax.spines['top'].set_position(('axes', 1.08))
    else:
        ax.grid()
    

# To tidy the plot up a little more, we can remove the depth labels in between each subplot / track and reduce the space between them.
#Hide tick labels on the y-axis 
for ax in [ax2,ax3]:
    plt.setp(ax.get_yticklabels(), visible=False)

#Reduce the space between each subplot
fig.subplots_adjust(wspace = 0.05)


plt.savefig('images/logplot.png', dpi=150)
```
