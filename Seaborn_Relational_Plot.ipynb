```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('data/Xeek_Well_15-9-15.csv', na_values=999) 
# Cualquier ocurrencia del valor 999 lo reemplazara como un valor faltante (NaN)
```

```python
df
```

### SCATTER PLOT WITH RELPLOT

```python
sns.scatterplot(data=df, x='NPHI', y='RHOB')
```

```python
sns.relplot(data=df, x='NPHI', y='RHOB')
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB')
g.set(ylim=(3,1.5))
```

### Customising the Scatter Plot

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='GR')
g.set(ylim=(3,1.5))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='GR', hue_norm=(0,150))
g.set(ylim=(3,1.5))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='CALI')
g.set(ylim=(3,1.5))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH')
g.set(ylim=(3,1.5))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH', style='LITH')
g.set(ylim=(3,1.5))
```

### Line Plot

```python
g = sns.relplot(data=df, x='DEPTH_MD', y='GR', kind='line')
g.set(ylim=(0,150))
```

```python
g = sns.relplot(data=df, x='DEPTH_MD', y='GR', kind='line', height=7, aspect=2)
g.set(ylim=(0,150))
```

### Splitting Data into Facets

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH', col='LITH')
g.set(ylim=(3,1.5 ))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH', col='LITH', col_wrap=3)
g.set(ylim=(3,1.5 ))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH', col='LITH', col_wrap=3, s=100)
g.set(ylim=(3,1.5 ))
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB', hue='LITH', col='LITH', row='GROUP') # 'GROUP' columna del Dataframe
g.set(ylim=(3,1.5 ))
```
